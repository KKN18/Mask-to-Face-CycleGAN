```
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/CycleGAN_Project/datasets/trainA/800.jpg')
```

```
cv2_imshow(image)
```

```
import numpy as np
kernel = np.array([[0, -1, 0], 
                       [-1, 5,-1],
                       [0, -1, 0]])
```

```
sharpened = cv2.filter2D(image, -1, kernel)
```

```
cv2_imshow(sharpened)
```

```
import glob
imgFiles = glob.glob('/content/drive/MyDrive/Colab Notebooks/CycleGAN_Project/datasets/trainA/*.jpg')
savePath = '/content/drive/MyDrive/Colab Notebooks/CycleGAN_Project/datasets/trainA2/{}.jpg'
kernel = np.array([[0, -1, 0], 
                       [-1, 5,-1],
                       [0, -1, 0]])
num = 0
for img in imgFiles:
  image = cv2.imread(img)
  sharpened = cv2.filter2D(image, -1, kernel)
  cv2.imwrite(savePath.format(num), sharpened)
  num += 1
```
