```
from PIL import Image
import glob
```

```
from os import listdir
from os.path import isfile, join
# from pylab import *
# from numpy import *

def getFolder(tPath, isFile = True):
    return [f for f in listdir(tPath) if isFile == isfile(join(tPath, f))]
```

```
tPath = 'datasets/testA'
sPath = 'datasets/tst'
tImgDic = getFolder(join(tPath))
total = len(glob.glob(tPath+'/*.jpg'))
num = 1
for image in tImgDic:
    img = Image.open(join(tPath, image))
    img = img.resize((256, 256))
    img.save(join(sPath, image))
    print('Complete : '+ str(num) + '/' + str(total))
    num += 1
```
